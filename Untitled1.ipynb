```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import numpy as np

import streamlit as st
import pandas as pd
import joblib
import numpy as np

# ---------------- Load Models ----------------
regression_model = joblib.load("xgboost_regression_model.joblib")
cluster_model = joblib.load("kmeans_cluster_model.joblib")
cluster_preprocessor = joblib.load("clustering_preprocessor.joblib")

st.title("ğŸ’ Diamond Price Prediction & Market Segmentation")

st.header("ğŸ”¹ Enter Diamond Details")

# ---------------- Input Fields ----------------
carat = st.number_input("Carat", min_value=0.1, step=0.01)
x = st.number_input("Length (x in mm)", min_value=0.1, step=0.01)
y = st.number_input("Width (y in mm)", min_value=0.1, step=0.01)
z = st.number_input("Depth (z in mm)", min_value=0.1, step=0.01)
depth = st.number_input("Total Depth (%)", min_value=40.0, step=0.1)
table = st.number_input("Table Width (%)", min_value=40.0, step=0.1)

cut = st.selectbox("Cut", ["Fair", "Good", "Very Good", "Premium", "Ideal"])
color = st.selectbox("Color", ["J", "I", "H", "G", "F", "E", "D"])
clarity = st.selectbox("Clarity", ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"])

carat_category = "Light" if carat < 0.5 else "Medium" if carat <= 1.5 else "Heavy"

# Derived features
Volume = x * y * z
price_per_carat = 0  # Placeholder (real price unknown before prediction)
dimension_ratio = (x + y) / (2 * z)

# ---------------- Create DataFrame ----------------
input_data = pd.DataFrame([{
    'carat': carat,
    'x': x,
    'y': y,
    'z': z,
    'depth': depth,
    'table': table,
    'Volume': Volume,
    'price_per_carat': price_per_carat,
    'dimension_ratio': dimension_ratio,
    'cut': cut,
    'color': color,
    'clarity': clarity,
    'carat_category': carat_category
}])

# ---------------- Prediction Buttons ----------------
if st.button("ğŸ’° Predict Price"):
    price_pred = regression_model.predict(input_data)[0]
    st.success(f"Predicted Diamond Price: â‚¹{price_pred:,.2f}")

if st.button("ğŸ’ Predict Market Segment"):
    try:
        # Transform input using preprocessor
        X_cluster = cluster_preprocessor.transform(input_data)

        # --- Auto-fix feature mismatch silently ---
        expected_features = cluster_model.n_features_in_
        if X_cluster.shape[1] < expected_features:
            diff = expected_features - X_cluster.shape[1]
            X_cluster = np.hstack([X_cluster, np.zeros((X_cluster.shape[0], diff))])

        # Predict cluster
        cluster_label = cluster_model.predict(X_cluster)[0]

        # Friendly names
        cluster_names = {
            0: "ğŸ’ Premium Heavy Diamonds",
            1: "ğŸ’ Mid-range Balanced Diamonds",
            2: "ğŸ’  Affordable Small Diamonds"
        }

        st.info(f"Predicted Segment: {cluster_names.get(cluster_label, 'Unknown Cluster')}")

    except Exception as e:
        st.error(f"Error during prediction: {e}")
```
